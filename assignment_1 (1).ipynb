# **Ques 3 - Compare and contrast mutable and immutable objects in Python with examples.**
answer -

**1 - Mutability**

Mutable: Can be changed after creation (e.g., lists, dictionaries).

Immutable: Cannot be changed once created (e.g., strings, tuples).

**2 -Memory Storage:**

Mutable: Stored as references to the memory location. Changes affect the original data.


Immutable: Stored as values. Any change creates a new memory location without altering the original value.

**3 -Function Arguments:**

Mutable: Changes within a function affect the original value (passed by reference).


Immutable: Changes within a function do not affect the original value (passed by value).

**Examples:**
Mutable: list, dict, set
Immutable: str, tuple, int, float


# **Ques 4 - Discuss the different types of operators in Python and provide examples of how they are used**

answer - In Python, operators perform operations on variables and values. They are categorized into several types:

1. **Arithmetic Operators**: Used for basic mathematical operations.
   - `+` (addition)
    `-` (subtraction)
     `*` (multiplication)
      `/` (division)
       `//` (floor division)
       `%` (modulus)
       `**` (exponentiation).
   - **Example**:
     
     a = 10
     b = 3
     print(a + b)  # Output: 13
     print(a / b) # Output: 3
     print(a - b) # Output: 10
     

2. **Comparison Operators**: Compare two values and return a boolean result.
   - `==`, `!=`, `>`, `<`, `>=`, `<=`.
   - **Example**:
   
     a = 5
     b = 10
     print(a == b)  # Output: False
     

3. **Logical Operators**: Combine conditional statements.
   - `and`, `or`, `not`.
   - **Example**:
   
     x = 5
     y = 10
     print(x < y and y < 20)  # Output: True
     

4. **Assignment Operators**: Assign values to variables.
   - `=`, `+=`, `-=`, `*=`, `/=`, etc.
   - **Example**:
     
     x = 5
     x += 3  # x = 8
     

5. **Membership Operators**: Check if a value is in a sequence.
   - `in`, `not in`.
   - **Example**:
   
     lst = [1, 2, 3]
     print(2 in lst)  # Output: True
     

6. **Identity Operators**: Compare object references.
   - `is`, `is not`.
   - **Example**:
     
     a = [1, 2]
     b = a
     print(a is b)  # Output: True
     

These operators enable efficient manipulation and comparison of data in Python.


# **Ques 5 - Explain the concept of type casting in Python with examples**.


answer -

**Type casting**  is the process of converting one data type into another. Python supports two types of type casting:

1. **Implicit Type Casting** (Automatic Conversion)
2. **Explicit Type Casting** (Manual Conversion)

### 1. **Implicit Type Casting**:
This happens automatically when Python converts a smaller data type to a larger one. For example, when an integer is added to a float, Python automatically converts the integer to a float to ensure there’s no data loss. This process is automatic and doesn't require any manual intervention.

**Example**:

a = 5           # Integer
b = 3.14        # Float
result = a + b  # Implicit conversion of 'a' to float
print(result)    # Output: 8.14

In this case, `a` (an integer) is automatically converted to a float before performing the addition.

### 2. **Explicit Type Casting**
This type of conversion requires the programmer to use built-in functions like `int()`, `float()`, `str()`, etc., to convert a variable from one type to another. It’s often used when you need to manually control the type conversion, especially when moving between incompatible types.

**Example 1** - Converting a float to an integer:

x = 9.8
y = int(x)  # Explicitly convert float to integer (loses decimal part)
print(y)    # Output: 9


**Example 2** - Converting a string to an integer:

s = "123"
num = int(s)  # Convert string to integer
print(num)    # Output: 123


- **Implicit casting** is automatic and occurs when there's no risk of data loss.
- **Explicit casting** requires the programmer's intervention and is done using functions like `int()`, `float()`, and `str()`.



# **Ques 6 - How do conditional statements work in Python? Illustrate with examples.**

Answer - Conditional statements in Python are used to make decisions in my code. They help me execute certain blocks of code based on whether a condition is true or false. The main types of conditional statements are if, elif, and else. In this assignment, I will explain how these work with examples.

if Statement: The if statement checks a condition. If the condition is true, the code block under it runs.
Example:


# Using the if statement
age = 18

if age >= 18:


    print("You are an adult.")  # This message will be printed

In this example, I check if age is 18 or older. Since this is true, the program prints "You are an adult."
     
else Statement: The else statement provides an alternative action if the if condition is false.
Example:


# Using the if and else statements
age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")  # This message will be printed

Here, since age is 16, the program doesn't print "You are an adult." Instead, it prints "You are a minor."
     
elif Statement: The elif statement stands for "else if." It allows me to check multiple conditions.
Example:


# Using if, elif, and else statements
grade = 85

if grade >= 90:
    print("You got an A.")
elif grade >= 80:
    print("You got a B.")  # This message will be printed
elif grade >= 70:
    print("You got a C.")
else:
    print("You need to improve.")

In this case, the program checks the grade. It finds that the grade is 85, so it prints "You got a B."
     


# **7.Describe the different types of loops in Python and their use cases with examples.**

Different Types of Loops in Python

In Python, loops are used to repeat a block of code multiple times. This is useful when I want to perform the same action several times.
The two main types of loops in Python are the for loop and the while loop.

In this assignment, I will explain each type of loop with examples.

for Loop: The for loop is used to go through items in a list, tuple, or string. It lets me run a block of code for each item in the sequence.
Example:


# Using a for loop
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print("I like", fruit)  # This prints each fruit

In this example, I have a list of fruits. The for loop goes through each fruit and prints "I like" followed by the fruit name.
Use Case: I use a for loop when I want to process items in a collection, like displaying all the names in a list.
     
while Loop: The while loop runs as long as a condition is true. This is helpful when I don't know how many times I need to repeat the code beforehand.
Example:

# Using a while loop
count = 0

while count < 5:
    print("Count is:", count)
    count += 1  # Increase count by 1

In this example, the while loop continues to run as long as count is less than 5. It prints the current count and then adds 1 to it each time.
Use Case: I use a while loop when I want to repeat something until a certain condition is met, like counting up to a number.
